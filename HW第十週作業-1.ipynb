### 桃園市各區結婚統計

```python
df=pd.read_csv("https://data.tycg.gov.tw/opendata/datalist/datasetMeta/download?id=f95d1a33-ec56-44c5-a9ec-3afba8157e39&rid=642ff4ab-3fdd-4d3d-be1a-311180f3aca3")
```

```python
df
```

```python
df["區域總數"]=df.loc[:,"一月":"十二月"].sum(axis=1)         #將各地區每個月分結婚對數加總
```

```python
df=df.sort_values(by=["區域總數"],ascending=False)          #將各地區依照總結婚對數排序
```

```python
df.loc[:,"一月":"區域總數"].sum(axis=0)      #算每個月分的結婚對數總和
```

```python
month_sum=pd.DataFrame([["月份總和",1652,915,1273,906,1645,1094,1058,602,1346,1565,1366,1626,15048]],columns=["月份區域別","一月","二月","三月","四月","五月","六月","七月","八月","九月","十月","十一月","十二月","區域總數"])
```

```python
original=df.loc[:]
```

```python
new_df=pd.concat([original,month_sum],ignore_index=True,sort=False)          #將每個月分的總和加入dataframe裡面
```

```python
new_df["占比"]=new_df.區域總數/15048      #算每個區結婚對數占整個桃園的比重
```
