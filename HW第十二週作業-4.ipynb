```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 第十二週：作業四（自己生成數據分類）

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
x,y=make_blobs(n_samples=1000,centers=4,n_features=2,random_state=0)
#生成100個樣本，內設四類
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired')
```

### 嘗試用MeanShift自動分類

```python
from sklearn.cluster import MeanShift
clf=MeanShift()
clf.fit(x)
plt.scatter(x[:,0],x[:,1],c=clf.predict(x),cmap='Paired')
```

```python
x0=np.arange(-5,5.2,0.02)
y0=np.arange(-3,11.5,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.2,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

他自動把所有數據分成兩類了，原本數據中藍色及紫色那兩類大多被分到了下面藍色的部分中，故嘗試調整精準度，再看他分得如何

```python
clf=MeanShift(bandwidth=0.9)
clf.fit(x)
plt.scatter(x[:,0],x[:,1],c=clf.predict(x),cmap='Paired')
```

```python
x0=np.arange(-5,5.2,0.02)
y0=np.arange(-3,11.5,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

結果又分得太多類了，再調整精準度看看

```python
clf=MeanShift(bandwidth=1.8)
clf.fit(x)
plt.scatter(x[:,0],x[:,1],c=clf.predict(x),cmap='Paired')
```

```python
x0=np.arange(-5,5.2,0.02)
y0=np.arange(-3,11.5,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

將精準度設到1.8似乎有比較接近原資料的分類，但只分成了三類

```python
clf=MeanShift(bandwidth=1.5)
clf.fit(x)
plt.scatter(x[:,0],x[:,1],c=clf.predict(x),cmap='Paired')
```

```python
x0=np.arange(-5,5.2,0.02)
y0=np.arange(-3,11.5,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)

plt.contourf(xm,ym,Z,alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

將精準度設在1.5和原資料的分類結果長得最像
