## 高雄港郵輪船期

```python
df=pd.read_csv('https://kh.twport.com.tw/Common/OdApi.ashx?n=AF8E381A5E86F15D&t=B9081DF834E98AC4')
```

```python
df
```

```python
df.drop("備註",axis=1,inplace=True)     #把備註欄位刪掉
```

```python
df=df[df.港口別==1].sort_values(by=["船舶總噸"],ascending=False)    #將在1號港口的船舶篩選出來，並依照船舶噸數排序
```

```python
df=df.fillna(0)      #把NaN都替換成0
```

```python
df=df[(df.前一港=="基隆")|(df.後一港=="基隆")]        #將有經過基隆的船隻篩選出來
```

```python
df
```

```python
df.index=range(1,10)         #把編號改成1到9
```
