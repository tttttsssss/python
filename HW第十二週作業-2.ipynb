# 第十二週:作業二 (酒的資料庫-非監督式學習)

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
x=wine.data
y=wine.target
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(x)
X=pca.transform(x)
#把所有變數x轉成二維資料
```

### 以KMeans非監督式學習自動分三類

在作業一時，使用監督式學習，放入全部或是部分相關性大的變數預測，結果都沒有到很好。於是在此利用非監督式學習，嘗試讓機器自動利用所有現有變數將數據分成三類(資料原本就分成三類)，來看結果如何。

```python
from sklearn.cluster import KMeans
```

```python
clf=KMeans(n_clusters=3)
```

```python
clf.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clf.labels_)    #非監督式學習讓KMeans自己分三類
```

```python
plt.scatter(X[:,0],X[:,1],c=y)      #資料內原本的分類
```

可以明顯看出原本的資料分類重疊性很高，很難用所有變數很好的分類。
而KMeans分出來的很清楚，但和原資料差異非常大。
