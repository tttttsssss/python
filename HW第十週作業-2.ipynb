## 2019國民年金核付人數及金額之統計表

```python
df=pd.read_csv('https://apiservice.mol.gov.tw/OdService/download/A17000000J-030208-r0q')
```

```python
df.head()
```

```python
df["每人核付金額"]=df.核付金額/df.核付人數    #加入每人核付金額欄位
```

```python
df
```

```python
df_new=df[df.年齡>100]        #把100歲以上的人挑出來
```

```python
df_new.sort_values(by=["年齡","每人核付金額"],ascending=False)      #再依照年齡及每人核付金額排序
```

```python
df_new[df_new.每人核付金額>3628]
```
